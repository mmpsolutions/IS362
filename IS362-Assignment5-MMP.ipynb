```python
# 3. On February 12th, 2013, which New York area airport had the windiest weather?
# Reading weather file from local directory
file2 = pd.read_csv('weather.csv')
file2.head()
```

```python
#Filtering to get the answer here
# EWR Airport had the the windiest weather in the NY metro area on 02/12/2013. 
file2[(file2['year']==2013)&(file2['month']==2)&(file2['day']==12)].sort_values(by='wind_gust', ascending=False).head()
```
